```python
import pandas as pd

# Criando um dicionário vazio para armazenar o estoque
estoque = {}

# Definindo a função para adicionar produtos ao estoque
def adicionarProduto():
    nomeProduto = input("Digite o nome do produto: ")
    qtdProduto = int(input("Digite a quantidade do produto: "))
    estoque[nomeProduto] = qtdProduto
    print(f"O produto {nomeProduto} foi adicionado ao estoque com sucesso!")

# Definindo a função para remover produtos do estoque
def removerProduto():
    nomeProduto = input("Digite o nome do produto que deseja remover: ")
    if nomeProduto in estoque:
        del estoque[nomeProduto]
        print(f"O produto {nomeProduto} foi removido do estoque com sucesso!")
    else:
        print(f"O produto {nomeProduto} não foi encontrado no estoque.")

# Definindo a função para exibir o estoque atual
def exibirEstoque():
    if len(estoque) == 0:
        print("O estoque está vazio.")
    else:
        print("Estoque atual:")
        print("="*50)
        for produto, quantidade in estoque.items():
            print(f"{produto} - {quantidade}")

# Executando o sistema
while True:
    print("="*50)
    print("="*50)
    print("Sistema de gerenciamento de estoque")
    print("1 - Adicionar produto")
    print("2 - Remover produto")
    print("3 - Exibir estoque")
    print("4 - Exportar para Excel")
    print("5 - Sair")
    print("="*50)
    print("="*50)
    opcao = int(input("Digite a opção desejada: "))
    
    if opcao == 1:
        adicionarProduto()
    elif opcao == 2:
        removerProduto()
    elif opcao == 3:
        exibirEstoque()
    elif opcao == 4:
        df = pd.DataFrame.from_dict(estoque, orient='index', columns=['Quantidade'])
        df.index.name = 'Produtos'
        df.to_excel('estoque.xlsx')
        print("Estoque exportado com sucesso para o arquivo 'estoque.xlsx'!")
    elif opcao == 5:
        print("Saindo do sistema...")
        break
    else:
        print("Opção inválida! Digite novamente.")
```
